```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(rc={"figure.figsize": (20, 10)})
plt.rc('figure', figsize=(20, 10))
import numpy as np
```

```python
path = "data/train/"
```

```python
action_cate8 = pd.read_csv(path + 'positive.csv', encoding='gbk', 
                                  parse_dates=['time_y'])
```

```python
action_cate8
```

```python
g = action_cate8.groupby(['user_id', 'sku_id'])
```

```python
all_time = []
for (u, s), group in g:
    type_g = group.groupby('type_y')
    
    for t, tg in type_g:
        type_time[t] = tg.sort_values(by='time_y', ascending=False)['time_y'].values[0]
    all_time.append(type_time)
#     break
#     cart = group[group.type==2]
#     buy = group[group.type==4]
#     last_cart = cart.sort_value('time_y', ascending=False)[0]['time_y']
#     last_buy.sort_value('time-y', ascending=True)[0]['time_y']
```
