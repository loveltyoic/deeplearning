```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(rc={"figure.figsize": (20, 10)})
plt.rc('figure', figsize=(20, 10))
import numpy as np
from utils import *
```

```python
path = "data/train/"
```

```python
action_cate8 = pd.read_csv(path + 'positive.csv', encoding='gbk', 
                                  parse_dates=['time_y'])
```

```python
action_cate8.groupby('user_id')
```

```python
g = action_cate8.groupby(['user_id', 'sku_id'])
```

```python
import datetime
all_time = []
for (u, s), group in g:
    type_g = group.groupby('type_y')
    arr = [u, s] + [None] * 12
    for t, tg in type_g:
        arr[t*2] = tg.sort_values('time_y', ascending=True)['time_y'].values[0]
        arr[t*2+1] = tg.sort_values('time_y', ascending=False)['time_y'].values[0]
    all_time.append(arr)

#     break
#     cart = group[group.type==2]
#     buy = group[group.type==4]
#     last_cart = cart.sort_value('time_y', ascending=False)[0]['time_y']
#     last_buy.sort_value('time-y', ascending=True)[0]['time_y']
```

```python
flag = ['start', 'end'] * 6
```

```python
positive_timetable = pd.DataFrame(np.array(all_time), 
                columns=['uid', 'sid'] + [str(i/2+1) + flag[i] for i in range(len(flag))])
positive_timetable.head(10)
```
